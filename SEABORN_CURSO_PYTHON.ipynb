O Seaborn é uma biblioteca de visualização de dados do Python baseado no Matplotlib. Ele provê uma interface de alto nível para construção de gráficos estatísticos atrativos e informativos.
[Site Seaborn org](https://seaborn.pydata.org/)

```python
import numpy as np
import pandas as pd
import seaborn as sns

# Definir tema
sns.set_theme(style='darkgrid')
```

```python
# Dados
Base_Dados = sns.load_dataset('tips')

# Varificar
Base_Dados.head()
```

```python
# Renomenado as colunas
Base_Dados.rename( columns={
    'total_bill' : 'Total_Conta',
    'tip' : 'Gorjeta',
    'sex' : 'Sexo',
    'smoker' : 'Fumante',
    'day' : 'Dia_Semana',
    'time' : 'Periodo',
    'size' : 'Pessoas_Mesa'
 }, inplace=True )
```

```python
Base_Dados.head()
```

**Gráfico relplot**

```python
# Gráfico relplot
sns.relplot( x='Total_Conta', y='Gorjeta', data=Base_Dados);
```

```python
# Gráfico relplot
# Passando outro parametro como classe no gráfico
sns.relplot( x='Total_Conta', y='Gorjeta', data=Base_Dados, hue='Sexo' );
```

```python
# Gráfico de Linha com 2 Eixos
sns.relplot( x='Total_Conta', y='Gorjeta', data=Base_Dados, kind='line' );
```

```python
# Gráfico de Linha com 2 Eixos
sns.relplot( x='Total_Conta', y='Gorjeta', data=Base_Dados,
            kind='line', hue='Fumante' );
```

**Gráfico histplot**

```python
sns.histplot( data=Base_Dados, x='Total_Conta')
```

```python
sns.histplot( data=Base_Dados, x='Total_Conta', hue='Fumante');
```

**Gráfico barplot**

```python
sns.barplot( data=Base_Dados, x='Sexo', y='Total_Conta', hue='Fumante');
```

**Gráfico pairplot**

```python
sns.pairplot( Base_Dados );
```

```python
sns.pairplot( Base_Dados, hue='Sexo' );
```

```python
sns.pairplot( Base_Dados, kind='kde' );
```

**Gráfico boxplot**

```python
sns.boxplot( data=Base_Dados, x='Dia_Semana', y='Total_Conta');
```

```python
sns.boxplot( data=Base_Dados, x='Dia_Semana', y='Total_Conta', hue='Sexo');
```

**Gráfico violinplot**

```python
sns.violinplot( data=Base_Dados, x='Dia_Semana', y='Total_Conta' );
```

**Gráfico swarmplot**

```python
sns.swarmplot(data=Base_Dados, x='Total_Conta', y='Sexo' );
```

**Gráfico jointplot**

```python
sns.jointplot( data=Base_Dados, x='Total_Conta', y='Gorjeta');
```
